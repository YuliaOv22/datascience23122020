Импортируйте библиотеки pandas и numpy.
Загрузите "Boston House Prices dataset" из встроенных наборов данных библиотеки sklearn. Создайте датафреймы X и Y из этих данных.

Разбейте эти датафреймы н а тренировочные (X_train, y_train) и тестовые (X_test, y_test) с помощью функции train_test_split так, чтобы размер тестовой выборки

Создайте модель линейной регрессии под названием lr с помощью класса LinearRegression из модуля sklearn.linear_model.

Обучите модель на тренировочных данных (используйте все признаки) и сделайте предсказание на тестовых.

Вычислите R2 полученных предказаний с помощью r2_score из модуля sklearn.metrics.

```python
import numpy as np
import pandas as pd
from sklearn.datasets import load_boston
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
```

```python
boston = load_boston()
```

```python
boston.keys()
```

```python
data = boston["data"]
feature_names = boston["feature_names"]
target = boston["target"]
```

```python
X = pd.DataFrame(data, columns=feature_names)
y = pd.DataFrame(target, columns=["price"])
```

```python
X.info()
```

```python
y.info()
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
lr = LinearRegression()
lr.fit(X_train, y_train)
y_pred = lr.predict(X_test)
```

```python
check_test = pd.DataFrame({
    "y_test": y_test["price"],
    "y_pred": y_pred.flatten()
})
check_test.head()
```

```python
r2_score(check_test["y_test"], check_test["y_pred"])
```

```python
X_corr = X.corr()
X_corr = np.round(X_corr, 2)
```

```python
high_corr = list(X_corr[(np.abs(X_corr["target"]) > 0.5) & (X_corr["target"] != 1)].T.columns)
high_corr
```

```python
X.drop("target", axis=1, inplace=True)
```

```python
high_corr_2 = X[high_corr] ** 2
high_corr_2
```

```python
high_corr_2.rename(columns=lambda x: x + "_2", inplace=True)
high_corr_2
```

```python
X = pd.concat([X, high_corr_2], axis=1)
X
```
